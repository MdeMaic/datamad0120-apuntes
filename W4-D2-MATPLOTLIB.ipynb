```python
x=np.arange(0,1,0.1)
y=np.arange(10)
plt.figure(figsize=(10,2))
print("X",x)
print("Y",y)

plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
#Scatter sirve para mostrar solo los valores que si pertenecen a tus valores de X e Y
plt.scatter(x,y)
```

```python
square_points = [
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
]

x = [m[0] for m in square_points]
y = [m[1] for m in square_points]

plt.figure(figsize =(6,6))
plt.plot(x,y,"o-")
```

```python
square_points = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
])
x = square_points[:,0]
y = square_points[:,1]

plt.figure(figsize =(6,6))
plt.plot(x,y,"o-")
```

```python
def square(L=1):
    square_points = np.array([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    
    return square_points

s = square(5)
plt.plot(s[:,0],s[:,1],"o-")
```

## Mirar apuntes de marc

```python
x = np.arange(10)
y = np.random.random(10)
print(x)
print(y)
#Primer atributo es el eje horizontal, el segundo es el vertical.
plt.bar(x,y)
```

```python
import pandas as pd

df = pd.DataFrame(np.random.random((2,10)))
print(df)
df.columns = [f"mes {i}" for i in range(10)]
df = df.T
df.plot.bar()
plt.title("HOLA")
plt.xlabel("Meses")
plt.ylabel("Ventas(€)")
```
