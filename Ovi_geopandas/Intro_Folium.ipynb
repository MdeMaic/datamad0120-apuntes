```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster

import pandas as pd
```

```python
bici = pd.read_csv('data/bici_clean.csv')
bici.head()
```

```python
# Create a base map
start_lat = 40.408561
start_lon = -3.6917665
heat_m = folium.Map(location=[start_lat, start_lon],tiles='cartodbpositron', zoom_start=15)
heat_m
```

```python
# Add a marker to the heat map
uni_lon, uni_lat = -3.698127, 40.4055128
Marker([uni_lat, uni_lon], icon=folium.Icon(color='red')).add_to(heat_m)
heat_m
```

```python
prado_lon, prado_lat = -3.6918889, 40.4131467
heat_m.add_child(Marker([prado_lat, prado_lon], icon=folium.Icon(color='green')))
heat_m
```

```python
# Create groups
bicis_manana = bici[bici.horario=='Mañana']

manana_group = folium.FeatureGroup(name="Mañana")

HeatMap(data=bicis_manana[['lat', 'lon']],radius=15).add_to(manana_group)
manana_group.add_to(heat_m)

heat_m
```

```python
bicis_tarde = bici[bici.horario=='Tarde']

tarde_group = folium.FeatureGroup(name="Tarde")

HeatMap(data=bicis_tarde[['lat', 'lon']],radius=15).add_to(tarde_group)
tarde_group.add_to(heat_m)

folium.LayerControl(collapsed=False).add_to(heat_m)

heat_m
```

```python
heat_m.save('data/primerito_mapa')
```
